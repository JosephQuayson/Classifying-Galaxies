```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator

from visualize import visualize_activations

from sklearn.model_selection import train_test_split
from utils import load_galaxy_data

import app

#loading Galaxy Data
input_data, labels = load_galaxy_data()

#Checking Data Shape
print("input data shape:"+ str(input_data.shape)+ ",", "label  data shape:"+
 str(labels.shape))


#2 divide the data into training and validation data
x_train, x_test, y_train, y_test = train_test_split(
    input_data, 
    labels, 
    test_size = 0.2,
    stratify = labels,
    random_state = 222
)

#3 preprocessing of the input

generate_training_data = ImageDataGenerator(
  rescale = 1 / 255,
)

#4  NumpyArrayIterators

training_iterator = generate_training_data.flow(
  x_train, 
  y_train,
  batch_size = 5
)

validation_iterator =generate_training_data.flow(
  x_test, 
  y_test, 
  batch_size =5
)

#5 building our model
model = tf.keras.Sequential()
model.add(tf.keras.Input(shape=(128, 128, 3)))
model.add(tf.keras.layers.Dense(4,activation="softmax"))



#7 
model = tf.keras.Sequential()
model.add(tf.keras.Input(shape=(128, 128, 3)))
model.add(tf.keras.layers.Conv2D(8, 3, strides=2, activation="relu")) 
model.add(tf.keras.layers.MaxPooling2D(
    pool_size=(2, 2), strides=(2,2)))
model.add(tf.keras.layers.Conv2D(8, 3, strides=2, activation="relu")) 
model.add(tf.keras.layers.MaxPooling2D(
    pool_size=(2,2), strides=(2,2)))
model.add(tf.keras.layers.Flatten())
model.add(tf.keras.layers.Dense(16, activation="relu"))
model.add(tf.keras.layers.Dense(4, activation="softmax"))


#6
model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),
    loss=tf.keras.losses.CategoricalCrossentropy(),
    metrics=[tf.keras.metrics.CategoricalAccuracy(),tf.keras.metrics.AUC()])
#8 Checking the summary of our model
model.summary()



#9 Training our model 
model.fit(
        training_iterator,
        steps_per_epoch=len(training_iterator.x)/5,epochs=8,
        validation_data=validation_iterator,
        validation_steps=len(validation_iterator.x)/5)

#12
visualize_activations(model,validation_iterator)
```
